## Anlysing Data

```python
plt.figure(figsize=(6, 6))
sns.barplot(data=d1, y=d1['State'].sort_values(ascending=True), x='Two Wheelers (Max power not exceeding 250 Watts)', palette='viridis')
plt.ylabel('State', fontsize=14, family='serif')
plt.xlabel('Number of 2 Wheeler Electic Vehicles', family='serif', fontsize=14, labelpad=10)
plt.title(label='Statewise 2 Wheeler Electric Vehicles in India', weight=200, family='serif', size=15, pad=12)
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.barplot(data=d1, y=d1['State'].sort_values(ascending=True), x='Three Wheelers (Category L5 as per CMVR)', palette='viridis')
plt.ylabel('State', fontsize=14, family='serif')
plt.xlabel('Number of 3 Wheeler Electic Vehicles', family='serif', fontsize=14, labelpad=10)
plt.title(label='Statewise 3 Wheeler Electric Vehicles in India', weight=200, family='serif', size=15, pad=12)
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.barplot(data=d1, y=d1['State'].sort_values(ascending=True), x='Passenger Cars (Category M1 as per CMVR)', palette='viridis')
plt.ylabel('State', fontsize=14, family='serif')
plt.xlabel('Number of 4 wheeler Electic Vehicles', family='serif', fontsize=14, labelpad=10)
plt.title(label='Statewise 4 Wheeler Electric Vehicles in India', weight=200, family='serif', size=15, pad=12)
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.barplot(data=d1, y=d1['State'].sort_values(ascending=True), x='Buses', palette='viridis')
plt.ylabel('State', fontsize=14, family='serif')
plt.xlabel('Number of Electic Buses', family='serif', fontsize=14, labelpad=10)
plt.title(label='Statewise Electric Buses in India', weight=200, family='serif', size=15, pad=12)
plt.show()
```

```python
x = d2['BodyStyle'].value_counts().plot.pie(radius=1.5, cmap='viridis', startangle=0, textprops=dict(family='serif'))
plt.pie(x=[2], radius=0.9, colors='white')
plt.title(label='Car body type preference of electric vehicles', family='serif', size=15, pad=60)
plt.ylabel('')
plt.show()
```

```python
x = d2['PowerTrain'].value_counts().plot.pie(radius=1.5, cmap='viridis', startangle=0, textprops=dict(family='serif'))
plt.pie(x=[1], radius=0.9, colors='white')
plt.title(label='Types of power train', family='serif', size=13, pad=60)
plt.ylabel('')
plt.show()
```
